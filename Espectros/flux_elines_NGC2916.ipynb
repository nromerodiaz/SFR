```python
#     Definimos la funcion que grafica la suma de las intensidades
def sumation(array):
    total_sum = np.zeros( ( len(array[1,:,:]), len(array[1,1,:])  ) )
    
    for i in range( len(array[1,:,:]) ):
        
        for j in range( len(array[1,1,:]) ):
            sum_pix = 1
            for k in range( len(array[:,:,:]) ):
                sum_pix += array[k,i,j] 
                
            total_sum[i,j] = sum_pix
      
    return total_sum
```

```python
#     Función de normalización del array 2d

def min_subtraction(array):
    return (array - amin(array) + 1)
```

```python
#     Funcion que encuentra el maximo valor de f_i y lo grafica
def find_max(array, data):
    arg_max = unravel_index(array.argmax(), array.shape)
    max_x = arg_max[1]
    max_y = arg_max[0]
    
    lambda_x = linspace(0,78,78)
    #aca por que me da distinto que en el notebook original? en el original es data[:,max_x,max_y]
    graph_max = plot( lambda_x , data[max_x, max_y, :] )
    xlabel('---')
    ylabel('Arbitrary units')
   
    return graph_max, max_x, max_y
```
