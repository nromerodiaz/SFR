```python
import numpy as np
import matplotlib.pyplot as plt
import fitsio
from scipy import optimize

%pylab inline
%matplotlib inline

import glob
```

# Definiciones de funciones útiles

```python
#     Declaramos los datos de una galaxia
# Esta funcion retorna un arreglo de arreglos que contienen:
# [ [flujo Ha], [flujo Hb], [razon Ha/Hb], [flujo OII], [EW OII] ]

def naming(galaxy_name):
    
    galax_data = []
    data = fitsio.read(galaxy_name)
    
    Ha_flux    = data[45,:,:]
    Hb_flux    = data[28,:,:]
    lineratio  = Ha_flux/Hb_flux
    OII_flux   = data[0,:,:]
    OII_EW     = data[153,:,:]
    
    galax_data.append(Ha_flux)
    galax_data.append(Hb_flux)
    galax_data.append(lineratio)
    galax_data.append(OII_flux)
    galax_data.append(OII_EW)
    
    galax_data = np.array(galax_data)
    
    return galax_data
```

```python
#     Definimos una funcion que grafica los parametros de una galaxia:
# Esta funcion retorna una visualizacion de: OII flux, Ha flux, Hb flux, log(Ha) vs log(Ha/Hb)

def param_plot(galaxy_data):
    
    pylab.rcParams['figure.figsize'] = (15,10)
    plt.figure()

    # Intensidad de la linea OII
    plt.subplot(221)
    plt.imshow(np.log10(galaxy_data[3]))
    plt.title('$\log_{10}(OII flux)$', fontweight='bold', fontsize='20')

    # Intensidad de la linea H_alpha
    plt.subplot(222)
    plt.imshow(np.log10(galaxy_data[0]))
    plt.title('$\log_{10}(Ha flux)$', fontweight='bold', fontsize='20')

    # Intensidad de la linea H_beta
    plt.subplot(223)
    plt.imshow(np.log10(galaxy_data[1]))
    plt.title('$\log_{10}(Hb flux)$', fontweight='bold', fontsize='20')

    # Razon de las intensidades Ha/Hb
    plt.subplot(224)
    plt.scatter(np.log10(galaxy_data[0]), np.log10(galaxy_data[2]), alpha=0.1)
    
    plt.xlabel('$\log_{10}(Ha_{flux})$', fontsize='20')
    plt.ylabel('$\log_{10}(Ha_{flux}/Hb_{flux})$', fontsize='20')
```

```python
#     Funcion que grafica la relacion log10(Ha) vs log10(Ha/Hb)

def lineratio_plot(galaxy_data):
    # Cortamos los datos
    ii = (galaxy_data[0]>0.05) & (galaxy_data[1]>0.0)
    
    pylab.rcParams['figure.figsize'] = (10,7)
    plt.xlabel('$\log_{10}(Ha_{flux})$', fontsize='25')
    plt.ylabel('$\log_{10}(Ha_{flux}/Hb_{flux})$', fontsize='25')

    plt.scatter(np.log10(galaxy_data[0][ii]), np.log10(galaxy_data[2][ii]), alpha=0.3)

    matplotlib.rc('xtick', labelsize=18) 
    matplotlib.rc('ytick', labelsize=18)
```

```python
#     Tratar los datos del eje x
# Al sacar las alturas de las barras de los histogramas,
# los datos del eje x consisten en los puntos donde 
# comienza & donde termina la barra del histograma, por
# lo cual el eje x es de dimension dim(x)=dim(y)+1
# Con esta funcion busco encontrar el punto medio 
# del grosor de cada barra y así tener dim(x)=dim(y).

# Es decir, por cada punto de altura de la barra (y)
# hay dos puntos de ancho de la barra (x) que indican el
# punto donde comienza la barra y donde termina. El punto 
# final de una barra es el punto inicial de la siguiente.

# Los datos del eje vertical no necesitan tratarse ya que
# tan solo son la altura de las barras del histograma

def treat_data(arr):
    new_x_data = []
    for i in range (1, len(arr)):
        x_point = 0.5*(arr[i]+arr[i-1])
        new_x_data.append(x_point)
    new_x_data = np.array(new_x_data)
    return new_x_data
```

```python
#     Histogramas
# Esta funcion nos permite ver la distribucion Ha/Hb 
# teniendo en cuenta la intensidad de Ha. Esto nos permite 
# ver la fluctuacion de Ha/Hb en terminos del SFR

# Esta funcion retorna el histograma de las 3 regiones del 
# espectro, ademas de las alturas de las bins y las fronteras 
# de las bins para las 3 regiones del expectro, faint (f),
# mid (m) y bright (b)

def lr_hist(galaxy_data):
    
    # Identificamos maximos y minimos de la intensidad Ha para 
    # poder dividir los datos en regiones faint, mid-range y bright
    ii = (galaxy_data[0]>0.1) & (galaxy_data[1]>0.0)
    mats = np.amax(galaxy_data[0][ii])
    mits = np.amin(galaxy_data[0][ii])
    
    # Ya teniendo los rangos, dividimos el espectro en tres partes
    # iguales el corte 1 (c1) siendo en el primer tercio y el 
    # corte 2 (c2) en el segundo tercio
    c1 = mits+(mats-mits)/3.0
    c2 = mits+(2.0*((mats-mits)/3.0))
    
    # Hacemos el histograma separando el espectro en regiones faint,
    # mid-range y bright
    # n_ :    indica la altura de las barras del histograma
    # bins_ : indica los puntos frontera de la barra del histograma 
    #         (2 puntos por barra, el punto final de una barra es 
    #         el inicial de la siguiente)
    ii = (galaxy_data[0]>0.1) & (galaxy_data[0]<c1) & (galaxy_data[1]>0.0) #faint
    n_faint, bins_f, histo  = plt.hist(np.log10(galaxy_data[2][ii]), bins=20, log=True, normed=True, color='b')
    
    # Tratamos los datos (f) horizontales del histograma
    bins_f = treat_data(bins_f)

    ii = (galaxy_data[0]>c1) & (galaxy_data[0]<c2) & (galaxy_data[1]>0.0)  #mid-range
    n_mid, bins_m, histo  = plt.hist(np.log10(galaxy_data[2][ii]), bins=4, log=True, normed=True, alpha=0.8, color='g')
    
    # Tratamos los datos (m) horizontales del histograma
    bins_m = treat_data(bins_m)

    ii = (galaxy_data[0]>c2) & (galaxy_data[1]>0.0)                        #bright
    n_bright, bins_b, histo  = plt.hist(np.log10(galaxy_data[2][ii]), bins=4, log=True, normed=True, alpha=0.4, color='r')
    
    # Tratamos los datos (b) horizontales del histograma
    bins_b = treat_data(bins_b)
    
    # Valor teorico Ha/Hb
    theoretical_ratio = 2.85
    
    pylab.rcParams['figure.figsize']=(10,7)
    plt.axvline(x=np.log10(theoretical_ratio), color='k', linestyle='--')
    plt.xlabel(r'$\log_{10}(Ha_{flux}/Hb_{flux})$', fontsize = '25' )

    matplotlib.rc('xtick', labelsize=20) 
    matplotlib.rc('ytick', labelsize=20)
    
    # Esta funcion retorna el histograma de las 3 regiones del espectro,
    # ademas de las alturas de las bins y las fronteras de las bins 
    # para las 3 regiones del expectro, faint (f), mid (m) y bright (b)
    
    return bins_f, n_faint, bins_m, n_mid, bins_b, n_bright
```

```python
#     Double Power Law
# Funcion deducida en [2], Forero-Romero & Djikstra

def double_power_law(M, P0, M0, alpha, gamma):
    
    return P0*( ( (M/M0)**alpha + (M/M0)**gamma )**-1 )
```

```python
#     Ajuste
# Ajuste de la distribucion del histograma
def fit(bin_x, bin_height):
    
    return optimize.curve_fit(double_power_law, bin_x, bin_height, p0=[1.15, 0.75, -0.13, 2.67])
```
