```python
#     Definimos la función que grafica la suma de las intensidades

def sumation(array):
    total_sum = np.zeros( ( len(array[1,:,:]), len(array[1,1,:])  ) )
    
    for i in range( len(array[1,:,:]) ):
        
        for j in range( len(array[1,1,:]) ):
            sum_pix = 1
            for k in range( len(array[:,:,:]) ):
                sum_pix += array[k,i,j] 
                
            total_sum[i,j] = sum_pix
      
    return total_sum
```

```python
#     PREGUNTARLE A JAIME
#     Función de normalización del array 2d

def min_subtraction(array):
    return (array - amin(array) + 1)
```

```python
#     Función que identifica el máximo valor de f_i y lo grafica

def find_max_plot(array, data):

    arg_max = unravel_index(array.argmax(), array.shape)
    max_x = arg_max[1]
    max_y = arg_max[0]
    
    lambda_x = linspace(3650, 4840, 1701)
    
    graph_max = plot( lambda_x , data[:, max_x, max_y] )
    xlabel('Angstrom')
    ylabel('Arbitrary units')
   
    return graph_max
    
    #     para el futuro:
    '''
    fig1 = plt.figure()
    ax = fig1.add_subplot(111)
    ax.plot(data[:,i,j], label="NGC0001")
    ax.legend(loc="best")
    ax.set_xlabel("Longitud de onda")
    ax.set_ylabel("Intensidad")
    ax.set_title("Espectros galacticos")

    plt.savefig("NGC0001_V1200.pdf", format = "pdf")
    
    return fig1
    '''
```
